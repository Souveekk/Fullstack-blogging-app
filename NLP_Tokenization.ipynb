```python
corpus = """
My name is Demetrius Johnson. I am a professional MMA fighter. I am 15 time world champion!, I have now retired and run a coaching school.
"""
```

```python
print(corpus)
```

```python
### Paragraph --> sentences
import nltk
from nltk.tokenize import sent_tokenize
nltk.download('punkt_tab')
documents=sent_tokenize(corpus)
```

```python
print(type(documents))
documents
```

```python
### Paragraphs --> words
### Sentences --> words

from nltk.tokenize import word_tokenize
words=word_tokenize(corpus)
print(type(words))
print(words)
```
